# 붓꽃 데이터에 결정 트리 적용 및 시각화

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

dt_clf = DecisionTreeClassifier(random_state=156)

iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target,
                                                   test_size=0.2, random_state=11)
dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(dt_clf, out_file='tree.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)
```

```python
import graphviz
with open('tree.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# tree.dot을 png파일로 변환 후 출력
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'decistion_tree.png', '-Gdpi=600'])

# 주피터에서 png 직접 출력
from IPython.display import Image
Image(filename = 'decistion_tree.png')
```

결정 트리의 하이퍼 파라미터
- max_depth = 결정 트리의 최대 트리깊이 제어
- min_samples_split = 자식 규칙 노드를 분할해서 만드는데 필요한 최소 샘플 데이터 수
- min_samples_leaf = 리프노드가 될 수 있는 최소 샘플 개수

```python
# 결정 트리 하이퍼 파라미터 튜닝
dt_clf = DecisionTreeClassifier(max_depth = 3, random_state=156)
dt_clf.fit(X_train, y_train)

export_graphviz(dt_clf, out_file='tree_maxDepth.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

import graphviz
with open('tree_maxDepth.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
dt_clf = DecisionTreeClassifier(max_depth = 4, random_state=156)
dt_clf.fit(X_train, y_train)

export_graphviz(dt_clf, out_file='tree_maxDepth2.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

import graphviz
with open('tree_maxDepth2.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
dt_clf = DecisionTreeClassifier(min_samples_split = 4, random_state=156)
dt_clf.fit(X_train, y_train)

export_graphviz(dt_clf, out_file='tree_minSamples.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

import graphviz
with open('tree_minSamples.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
dt_clf = DecisionTreeClassifier(min_samples_split = 5, random_state=156)
dt_clf.fit(X_train, y_train)

export_graphviz(dt_clf, out_file='tree_minSamples.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

import graphviz
with open('tree_minSamples.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
dt_clf = DecisionTreeClassifier(min_samples_leaf = 4, random_state=156)
dt_clf.fit(X_train, y_train)

export_graphviz(dt_clf, out_file='tree_minleaf.dot', class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

import graphviz
with open('tree_minleaf.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```
